## Convex or concave?

```python
a = Point('A', 1, 0)
b = Point('B', 0, 0)
c = Point('C', 0.5, 0.2)
d = Point('D', 0.2, 0.9)
f = Point('F', 0.2, 0.2)

e = Point('E', 0.7, 0.6)

fig, ax =  get_fig_base()
polygon_convex = Polygon([d,e,c,f])
polygon_convex.draw(fig, ax)
print(polygon_convex.is_convex())

fig, ax =  get_fig_base()
polygon_concave = Polygon([a,d,e,b,c])
polygon_concave.draw(fig, ax, pointcolor='black', linecolor='blue')
print(polygon_concave.is_convex())
```

## Sample polygons

```python
number_of_samples = 3

for i in range(number_of_samples):    
    n = np.random.randint(4, 8)
    new_coordinates = np.random.uniform(size = (n, 2))
    new_points = [Point("p" + str(i+1), x,y) for i, (x,y) in enumerate(new_coordinates)]
    
    fig, ax = get_fig_base()
    polygon = Polygon(new_points)
    polygon.draw(fig, ax)
```
