## Simple unit tests for class Line

```python
p1 = Point('p1', 0, 0.2)
p2 = Point('p2', 1, 0.7)
line = Line.create_line_from_points(p1, p2)
line
assert np.isclose(line.intercept, 0.2)
assert np.isclose(line.slope, 0.5)
```

## Convex or concave?

```python
a = Point('A', 1, 0)
b = Point('B', 0, 0)
c = Point('C', 0.5, 0.2)
d = Point('D', 0.2, 0.9)
f = Point('F', 0.2, 0.2)

e = Point('E', 0.7, 0.6)

fig, ax =  get_fig_base()
polygon_convex = Polygon([d,e,c,f])
polygon_convex.draw(fig, ax)
print(polygon_convex.is_convex())

fig, ax =  get_fig_base()
polygon_concave = Polygon([a,d,e,b,c])
polygon_concave.draw(fig, ax, pointcolor='black', linecolor='blue')
print(polygon_concave.is_convex())
```

```python
a = Point('A', 1, 0)
b = Point('B', 0, 0)
c = Point('C', 0.5, 0.2)
d = Point('D', 0.2, 0.9)
f = Point('F', 0.2, 0.2)
e = Point('E', 0.7, 0.6)
poly = Polygon([a,d,e,b,c])


fig, ax =  get_fig_base()


color = ['pink', 'blue', 'green', 'orange', 'magenta', 'gray']

coordinates = np.random.uniform(size=(5000, 2))
points = [Point('', x, y) for i, (x,y) in enumerate(coordinates)]


left_border = min([p.x for p in poly.vertices])
right_border = max([p.x for p in poly.vertices])
bottom_border = min([p.y for p in poly.vertices])
top_border = max([p.y for p in poly.vertices])
    
poly.draw(fig, ax)

for pp in points:
    drawn = False
    counter = 0
    

    if left_border<pp.x<right_border and bottom_border<pp.y<top_border:   
        
        for i, edge in enumerate(poly.edges):

            if edge.is_point_in_line(pp):
                pp.draw_point(fig, ax, color='red')
                break

            line = edge.get_line_to_point(pp)

            if edge.is_point_in_line(line.ending_point):
                counter +=1

                inner_counter = 0

                for interior_edge in poly.edges:
                    #innner_counter+=1
                    CP = line.get_crossing_point(interior_edge)

                    if interior_edge.is_point_in_line(CP) and line.is_point_in_line(CP) and \
                            line.starting_point != CP and line.ending_point!= CP:
                        inner_counter+=1

                if inner_counter%2==1:
                    pp.draw_point(fig, ax, color='brown', s=5)
                    drawn = True

        if drawn==False:

            if counter == 1:
                pp.draw_point(fig, ax, color='orange', s=5)
            else:     
                pp.draw_point(fig, ax, color='green', s=5)
    else:
        pp.draw_point(fig, ax, color='magenta', s=5)
```
